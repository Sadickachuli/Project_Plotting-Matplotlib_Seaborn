```python
# YOUR CODE HERE
# A histogram that depicts the distribution of carbon Dioxide
co2_emissions = df['co2']

# Width and height
plt.figure(figsize=(20, 10))

# histogram style
plt.style.use('fivethirtyeight')

# With the bins, it seemed like anything above 10 was a bit too much.
plt.hist(co2_emissions, bins=10, color='orange', edgecolor='black')

# Labeling and showing the histogram
plt.xlabel('CO2 Emission')
plt.ylabel('Frequency')
plt.title('Distribution of CO2 Emission Histogram')
plt.tight_layout()
plt.show()
```

**Interpretation**

This distribution simply shows the differences between the emission levels and the concentration of emissions within particular ranges. The x-axis helps to identify the range of emissions, while the y-axis shows the number of vehicles that emit within each range.



```python
# YOUR CODE HERE

# Plot histogram for city fuel efficiency
plt.figure(figsize=(12, 10))
plt.style.use('fivethirtyeight')
plt.hist(df['city'], bins=20, color='red', edgecolor='black')
plt.xlabel('City Fuel Efficiency')
plt.ylabel('Frequency')
plt.title('City Histogram')

plt.tight_layout()
plt.show()
```

**Interpretation**

I can observe that the majority of the vehicles in the dataset fall into a particular range of values for their city fuel efficiency.
The distribution appears to be slightly moved to the right, meaning there are relatively fewer vehicles with exceptionally high city fuel efficiency than there are cars with lower fuel efficiency.
The city fuel economy values that occur most frequently fall into a specific range, suggesting that all of the vehicles in the sample may be subject to comparable norms or patterns when driving in cities.



```python
# YOUR CODE HERE
plt.hist(df['highway'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Highway Fuel Efficiency')
plt.ylabel('Frequency')
plt.title('Distribution of Highway Fuel Efficiency')


plt.tight_layout()
plt.show()
```

**Interpretation**

Compared to city fuel efficiency, there still seem to be more values spread out across the peaks in frequency that indicate common values for highway fuel efficiency.
This means that a wider range of fuel efficiency scores on highways may result from distinct factors influencing highway fuel efficiency as opposed to driving circumstances in cities.

```python
#create a heat map of the data

import seaborn as sns
# select only numbers(integers/floats) and calculate the correlation_matrix
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(14, 10))

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Interpretation**

The heatmap indicates a strong positive correlation between the city and highway fuel efficiency measurements and the combined fuel efficiency (comb) and greenhouse gas score (ghgScore) which means that cars with higher fuel efficiency generally have higher greenhouse gas scores under both city and highway driving. Additionally, there is a strong negative correlation between CO2 emissions and highway and comb fuel efficiency. This makes sense since cars with lower fuel consumption typically emit less CO2.
